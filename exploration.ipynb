```python
import glob

all_docs = glob.glob('D:/ML_Data/UofTGeneImage/VarientxUofTHacks Gene Resource/*.jpg') + glob.glob('D:/ML_Data/UofTGeneImage/VarientxUofTHacks Gene Resource/*.png')
```

```python
len(all_docs)
```

```python
import re
file_names2 = []

for i in range(len(all_docs)):
    result = re.search('\\\\(.*)....', all_docs[i])
    file_names2.append(result.group(1))
```

```python
def output_text(dir):
    return pytesseract.image_to_string(dir)
```

```python
for i in range(len(all_docs)):
    text_file = open(f"output_model_text/{file_names2[i]}.txt","w+")
    text_file.write(output_text(all_docs[i]))
    text_file.close()
```
